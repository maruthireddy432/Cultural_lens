## Heritage_data

```python
heritage_data=df[:50]
heritage_data.head()
```

```python
heritage_data.drop(columns=['# Practicing Artisans'],inplace=True)
heritage_data.columns=['Art form','Region','Endangered','Coordinates','Festivals/Events','UNESCO Listed']
```

```python
heritage_data=pd.concat([heritage_data,heritage_data['Coordinates'].str.split(',',expand=True)],axis=1)
heritage_data.rename(columns={0: 'latitude', 1: 'longitude'}, inplace=True)
heritage_data['Endangered'] = heritage_data['Endangered'].str.replace(r'Yes.*', 'Yes', regex=True)
heritage_data.drop(columns=['Coordinates'],inplace=True)
heritage_data['latitude']=heritage_data['latitude'].str.replace(r'~','',regex=True)
heritage_data['longitude']=heritage_data['longitude'].str.replace(r'[a-z,A-Z,()]','',regex=True)
heritage_data['latitude'].astype(float)
heritage_data['longitude'].astype(float)
```

```python
heritage_data.head()
```

```python
heritage_data.to_csv('heritage_data.csv',index=False)
```
